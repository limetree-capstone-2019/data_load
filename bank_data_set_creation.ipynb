```python
sql_bank_open_close = """WITH bank AS(
    SELECT LEFT(ZIP_CD,5) as ZIP, LEFT(DT_EXIST_TERM,4) as branch_Close, LEFT(DT_OPEN,4) 
                        as branch_open, STATE_ABBR_NM as State FROM FFIEC_BRANCHES WHERE DOMESTIC_IND='Y'
)

SELECT a.year, a.STATE, a.ZIP, ifNull(num_open,0) as bank_open, 
       ifNull(num_close,0) as bank_close, (bank_open-bank_close) as bank_net, 
       (CASE WHEN bank_net > 0 THEN 1 WHEN bank_net = 0 THEN 0 ELSE -1 END) as GROUND_TRUTH 
FROM 
      
--Left Join base data
                    (SELECT year, zip, state from (select seq4() + 2004 as year
                                 from table(generator(rowcount => 15))
                                ) 
                     FULL OUTER JOIN bank
                     Group BY year, state, zip
                     ) AS a
                     
--Bank data
                     LEFT OUTER JOIN (
                       (SELECT ZIP,STATE,2004 as year_b, count(ZIP) as num_close
                        from bank 
                        where branch_Close = 2004
                        Group BY ZIP,STATE)
                        UNION             
                        (SELECT ZIP,STATE,2005 as year_b, count(ZIP) as num_close
                        from bank 
                        where branch_Close = 2005
                        Group BY ZIP,STATE)   
                        UNION             
                        (SELECT ZIP,STATE,2006 as year_b, count(ZIP) as num_close
                        from bank 
                        where branch_Close = 2006
                        Group BY ZIP,STATE) 
                        UNION             
                        (SELECT ZIP,STATE,2007 as year_b, count(ZIP) as num_close
                        from bank 
                        where branch_Close = 2007
                        Group BY ZIP,STATE) 
                        UNION             
                        (SELECT ZIP,STATE,2008 as year_b, count(ZIP) as num_close
                        from bank 
                        where branch_Close = 2008                         
                        Group BY ZIP,STATE) 
                        UNION             
                        (SELECT ZIP,STATE,2009 as year_b, count(ZIP) as num_close
                        from bank 
                        where branch_Close = 2009
                        Group BY ZIP,STATE) 
                        UNION             
                        (SELECT ZIP,STATE,2010 as year_b, count(ZIP) as num_close
                        from bank 
                        where branch_Close = 2010
                        Group BY ZIP,STATE) 
                        UNION             
                        (SELECT ZIP,STATE,2011 as year_b, count(ZIP) as num_close
                        from bank 
                        where branch_Close = 2011
                        Group BY ZIP,STATE) 
                        UNION             
                        (SELECT ZIP,STATE,2012 as year_b, count(ZIP) as num_close
                        from bank 
                        where branch_Close = 2012
                        Group BY ZIP,STATE) 
                        UNION             
                        (SELECT ZIP,STATE,2013 as year_b, count(ZIP) as num_close
                        from bank 
                        where branch_Close = 2013
                        Group BY ZIP,STATE) 
                        UNION             
                        (SELECT ZIP,STATE,2014 as year_b, count(ZIP) as num_close
                        from bank 
                        where branch_Close = 2014
                        Group BY ZIP,STATE) 
                        UNION             
                        (SELECT ZIP,STATE,2015 as year_b, count(ZIP) as num_close
                        from bank 
                        where branch_Close = 2015
                        Group BY ZIP,STATE) 
                        UNION             
                        (SELECT ZIP,STATE,2016 as year_b, count(ZIP) as num_close
                        from bank 
                        where branch_Close = 2016
                        Group BY ZIP,STATE) 
                        UNION             
                        (SELECT ZIP,STATE,2017 as year_b, count(ZIP) as num_close
                        from bank 
                        where branch_Close = 2017
                        Group BY ZIP,STATE) 
                        UNION             
                        (SELECT ZIP,STATE,2018 as year_b, count(ZIP) as num_close
                        from bank 
                        where branch_Close = 2018
                        Group BY ZIP,STATE) 
                                     ) AS b
                     ON a.ZIP = b.ZIP and a.STATE = b.STATE and a.year = b.year_b

                     LEFT OUTER JOIN(
                         (SELECT ZIP, STATE, 2004 as year_c, count(ZIP) as num_open
                         from bank 
                         where branch_open = 2004
                         Group BY ZIP, STATE)
                         UNION
                         (SELECT ZIP, STATE, 2005 as year_c, count(ZIP) as num_open
                         from bank 
                         where branch_open = 2005
                         Group BY ZIP, STATE)
                         UNION
                         (SELECT ZIP, STATE, 2006 as year_c, count(ZIP) as num_open
                         from bank 
                         where branch_open = 2006
                         Group BY ZIP, STATE)
                         UNION
                         (SELECT ZIP, STATE, 2007 as year_c, count(ZIP) as num_open
                         from bank 
                         where branch_open = 2007
                         Group BY ZIP, STATE)
                         UNION
                         (SELECT ZIP, STATE, 2008 as year_c, count(ZIP) as num_open
                         from bank 
                         where branch_open = 2008
                         Group BY ZIP, STATE)
                         UNION
                         (SELECT ZIP, STATE, 2009 as year_c, count(ZIP) as num_open
                         from bank 
                         where branch_open = 2009
                         Group BY ZIP, STATE)
                         UNION
                         (SELECT ZIP, STATE, 2010 as year_c, count(ZIP) as num_open
                         from bank 
                         where branch_open = 2010
                         Group BY ZIP, STATE)
                         UNION
                         (SELECT ZIP, STATE, 2011 as year_c, count(ZIP) as num_open
                         from bank 
                         where branch_open = 2011
                         Group BY ZIP, STATE)
                         UNION
                         (SELECT ZIP, STATE, 2012 as year_c, count(ZIP) as num_open
                         from bank 
                         where branch_open = 2012
                         Group BY ZIP, STATE)
                         UNION
                         (SELECT ZIP, STATE, 2013 as year_c, count(ZIP) as num_open
                         from bank 
                         where branch_open = 2013
                         Group BY ZIP, STATE)
                         UNION
                         (SELECT ZIP, STATE, 2014 as year_c, count(ZIP) as num_open
                         from bank 
                         where branch_open = 2014
                         Group BY ZIP, STATE)
                         UNION
                         (SELECT ZIP, STATE, 2015 as year_c, count(ZIP) as num_open
                         from bank 
                         where branch_open = 2015
                         Group BY ZIP, STATE)
                         UNION
                         (SELECT ZIP, STATE, 2016 as year_c, count(ZIP) as num_open
                         from bank 
                         where branch_open = 2016
                         Group BY ZIP, STATE)
                         UNION
                         (SELECT ZIP, STATE, 2017 as year_c, count(ZIP) as num_open
                         from bank 
                         where branch_open = 2017
                         Group BY ZIP, STATE)
                         UNION
                         (SELECT ZIP, STATE, 2018 as year_c, count(ZIP) as num_open
                         from bank 
                         where branch_open = 2018
                         Group BY ZIP, STATE)
                                     ) AS c
                     ON a.ZIP = c.ZIP and a.STATE = c.STATE and a.year = c.year_c
                     

Where a.ZIP != 0 
ORDER BY a.YEAR, a.STATE, a.ZIP

"""
```

```python
# ++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++
# [model creation]
# 
 
from sklearn.datasets import load_iris
from sklearn import tree
import csv

# drop year, state, zip
bank_vs_demo = bank_vs_demo.drop(['index','year','state','zip','bank_open','bank_close',
                                                            'bank_net'], axis=1)
data_list = bank_vs_demo.drop('ground_truth', axis=1)
label_list = bank_vs_demo['ground_truth']  

```

```python
from sklearn.model_selection import train_test_split

#Create train and test data. test_size 0.3 means 30% of data will be test data.
#change dataframe to matrix - data_list and label-list
train_data, test_data, train_label, test_label = train_test_split(data_list.as_matrix(), label_list.as_matrix(), test_size=0.3, random_state=None )
```

```python
#option1 decision tree classifier model
clf = tree.DecisionTreeClassifier(max_depth = 3)
clf = clf.fit(train_data, train_label)

# create prediction
predicted = clf.predict(test_data)

# print out predict and correct label
print('=============== predicted ===============')
print(predicted)
print('============== correct_ans (label) ==============')
print(test_label)
print('=============== accuracy =================')
print(sum(predicted == test_label) / len(test_label))

```

```python
#option2 support vector machine

from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix 
import matplotlib.pyplot as plt
import seaborn as sns

model_svc = SVC(kernel='linear', random_state=None)
model_svc.fit(train_data, train_label)

# create predict label
pred_test = model_svc.predict(test_data)

# create confusion matrix
results = confusion_matrix(test_label, pred_test)

# print the accuracy score on the test data
print('Accuracy Score :',accuracy_score(test_label, pred_test) )

# create label
labels = ['Open','Constant','Close']
```

```python
# display the heatmap of confusion matrix on the test data
fig, ax = plt.subplots(figsize=(10, 5)) 
ax.set_title('Confusion Matrix - Test Data')
sns.heatmap(results, annot =True, fmt="d", xticklabels=labels, yticklabels=labels)

plt.xlabel('Predicted Labels')
plt.ylabel('Groundtruth')

plt.show()
```
